```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set(style="white", color_codes = True)
# style ="white"  is to set the style of the plot to white , background will be white
#color_codes = True  telling  seaborn to use color coes when mapping categorical variables to colors
```

```python
iris = pd.read_csv("Iris.csv")
```

```python
print(iris.head())
# print the first five rows 
```

```python
print(iris["Species"].value_counts())
# print out counts of each species
```

```python
iris.plot(kind="scatter",x="SepalLengthCm",y="SepalWidthCm")
plt.show()
# plotting from x and y
```

```python
iris.plot(kind="scatter",x="SepalLengthCm",y="SepalWidthCm",color="green",s=70)
plt.show()
# s is size and color = change color of the plt plot
```

```python
sns.jointplot(x="SepalLengthCm",y="SepalWidthCm",data=iris, size = 5)
plt.show()
# bivariate scatterplot and univariate scatterplot
```

```python
sns.FacetGrid(iris, hue="Species", height=5) \
    .map(plt.scatter, "SepalLengthCm", "SepalWidthCm") \
    .add_legend()
plt.show()
# add_legend is to enable it to display the center right legend
```

```python
sns.boxplot(x="Species", y="PetalLengthCm", data=iris)
plt.show()
#boxplot
```

```python
ax = sns.boxplot(x="Species",y="PetalLengthCm", data =iris)

ax= sns.stripplot(x="Species",y="PetalLengthCm", data =iris,\
                jitter = True, edgecolor="gray")
plt.show()
```

```python
sns.violinplot(x="Species", y="PetalLengthCm", data=iris)
plt.show()
```

```python
sns.pairplot(iris.drop("Id",axis=1), hue="Species", height=3 )
plt.show()
```
