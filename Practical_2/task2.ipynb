```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
print(titanic.head())
```

```python
print(titanic.isnull().sum())
#check for missing values
```

```python
sns.countplot(x="Survived", data=titanic, hue = "Sex")
plt.show()
# check for survival between Male and Female
```

```python
sns.countplot(x="Survived", data=titanic , hue ='Pclass')
plt.show()
#check survival based on Pclass
```

```python
titanic['Age'].dropna().plot.hist(bins=30)
plt.show()
# we specified 30 bins = data will be divided into 30 intervals or histogram will have 30 bars
```

```python
sns.countplot(x="SibSp", data=titanic)
plt.show()
#check how many passengers with siblings
```

```python
titanic['Fare'].hist(bins=40, figsize=(10,4))
plt.show()
```

```python
sns.catplot(x="Embarked", data=titanic, hue="Pclass", kind='count')
plt.show()
```
