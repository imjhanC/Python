```python
##Confusion Matrix and AUC with logistic regression
# measure the performance
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_curve , auc
import matplotlib.pyplot as plt
```

```python
#generate abinary classfication dataset
X, y = make_classification (n_samples=1000, n_features=20, n_classes=2, random_state =42)
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state =42)

#Train a LOGISTIC REGRESSION CLASSIFIER
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
accuracy = accuracy_score (y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')
```

```python
# recall looks ar the actual "postive" values in the dataset and answers , how many did we correctly predict as positive
recall = recall_score(y_test, y_pred)
print(f'Recall: {recall:.2f}')
```

```python
#receiver operating characteristic ROC 
fpr, tpr, thresholds =roc_curve(y_test, clf.predict_proba(X_test)[:,1])
plt.plot(fpr,tpr, label='ROC curve')
plt.plot([0,1],[0,1], 'k--') #random predictions curve
plt.xlabel('False Positive RAte or (1 - Specificity)')
plt.ylabel('True Positive RAte or Sensitivity')
plt.title('Receiver Operating Characteristic (ROC)')
plt.legend(loc="lower right")
plt.show()
```

```python
#AUC area under the curve
roc_auc =auc(fpr,tpr)
print(f'AUC:{roc_auc:.2f}')
```
